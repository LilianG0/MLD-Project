```python
#!/bin/python3

#########################
######### ROUTINE TO PLOT THE SEASONAL MEAN OF ANY VARIABLE OVER AN ALREADY 
######### SELECTED REGION.
#########################


import cartopy
import cartopy.feature as cfeat
import cartopy.crs as ccrs
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import xarray as xr
import numpy as np

import datetime
import pandas

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
#directory = '/net/alpha/exports/sciences/data/LPO_ISAS/ANA_ISAS15/fld2D/'
directory1 = '/home/lgarcia/Documents/data_ARMOR/'
file1 = 'A-GUL_'
file2 = 'I-GUL.nc'

name_fig = 'MLD_' + 'GUL' + '.pdf'

datasets = []
yrs = [i for i in range(2006,2018,1)]

for i,year in enumerate(yrs):
    y = str(year)
    d = xr.open_dataset(directory1 + file1 + y + '.nc', autoclose=True)
    datasets.append(d)

combined = xr.concat(datasets, dim='time')

d2 = xr.open_dataset(directory1 + file2, autoclose=True)


mld1 = combined.mlotst[:,0,0]
mld2 = d2.MLDP[:,0,0,0]

%matplotlib notebook
mld1.plot.line(figsize=(10, 5))
mld2.plot.line()
plt.title('Gulf Stream MLD')
plt.plot(figsize=(15, 7.5))
#plt.savefig(name_fig, bbox_inches='tight')



```
